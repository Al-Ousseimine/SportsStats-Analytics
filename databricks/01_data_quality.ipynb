# Olympic Dataset

## Dataset description

SportsStats
 (ensemble de données sur les Jeux olympiques - 120 ans de données)

SportsStats est une société d'analyse sportive qui travaille en partenariat avec les journaux locaux et les entraîneurs personnels d'élite afin de fournir des informations "intéressantes" pour aider leurs partenaires. Les informations peuvent être des modèles/tendances mettant en évidence certains groupes/événements/pays, etc. dans le but de développer un article d'actualité ou de découvrir des informations clés sur la santé.

## Data loading
  Le jeu de données comporte deux tables: **athlete_events** qui contient des données liées aux événements sportifs des athlètes, y compris des détails sur chaque athlète comme leur nom, âge, taille, poids, équipe et les jeux auxquels ils participent et **noc_regions**  qui contient des informations sur différentes régions du NOC (Centre d'Opérations Réseau). Il inclut des détails tels que l'identifiant spécifique du NOC, le nom de la région associée et toutes les notes pertinentes. 
  
  Pour la **table athlete_events** on a les colonnes suivantes:
-  ID de type bigint
-   Name de type string,
-   Age  enregistré en type string mais devra être converti en numérique, 
-   Height  enregistré en type string mais devra être converti en numérique ,
-   weight  enregistré en type string mais devra être converti en numérique ,
-   Team de type string,
-   NOC de type string,
-   Games de type string,
-   Year de type bigint,
-   Season de type string,
-   City de type string,
-   Sport de type string,
-   Event de type string,
-   Medal de type string. 

Pour la **table noc_regions** on a les colonnes suivantes:
- NOC de type string,
- region de type string,
- notes de type string

```sql
SELECT 
       sum(CASE WHEN ID IS NULL THEN 1 ELSE 0 END) AS missing_ID,
       sum(CASE WHEN Name IS NULL OR Name = '' OR Name = 'NA' THEN 1 ELSE 0 END) AS missing_Name,
       sum(CASE WHEN Sex IS NULL OR Sex = '' OR Sex = 'NA' THEN 1 ELSE 0 END) AS missing_Sex,
       sum(CASE WHEN Age IS NULL OR Age = '' OR Age = 'NA' THEN 1 ELSE 0 END) AS missing_Age,
       sum(CASE WHEN Height IS NULL OR Height = '' OR Height = 'NA' THEN 1 ELSE 0 END) AS missing_Height,
       sum(CASE WHEN Weight IS NULL OR Weight = '' OR Weight = 'NA' THEN 1 ELSE 0 END) AS missing_Weight,
       sum(CASE WHEN Team IS NULL OR Team = '' OR Team = 'NA' THEN 1 ELSE 0 END) AS missing_Team,
       sum(CASE WHEN NOC IS NULL OR NOC = '' OR NOC = 'NA' THEN 1 ELSE 0 END) AS missing_NOC,
       sum(CASE WHEN Games IS NULL OR Games = '' OR Games = 'NA' THEN 1 ELSE 0 END) AS missing_Games,
       sum(CASE WHEN Year IS NULL THEN 1 ELSE 0 END) AS missing_Year,
       sum(CASE WHEN Season IS NULL OR Season = '' OR Season = 'NA' THEN 1 ELSE 0 END) AS missing_Season,
       sum(CASE WHEN City IS NULL OR City = '' OR City = 'NA' THEN 1 ELSE 0 END) AS missing_City,
       sum(CASE WHEN Sport IS NULL OR Sport = '' OR Sport = 'NA' THEN 1 ELSE 0 END) AS missing_Sport,
       sum(CASE WHEN Event IS NULL OR Event = '' OR Event = 'NA' THEN 1 ELSE 0 END) AS missing_Event,
       sum(CASE WHEN Medal IS NULL OR Medal = '' OR Medal = 'NA' THEN 1 ELSE 0 END) AS missing_Medal,
       
FROM athlete_events


```

```sql
WITH counts AS (
    SELECT 
        count(*) AS total_lignes_athlete_events,
        sum(CASE WHEN Age IS NULL OR Age = '' OR Age = 'NA' THEN 1 ELSE 0 END) AS missing_Age,
        sum(CASE WHEN Height IS NULL OR Height = '' OR Height = 'NA' THEN 1 ELSE 0 END) AS missing_Height,
        sum(CASE WHEN Weight IS NULL OR Weight = '' OR Weight = 'NA' THEN 1 ELSE 0 END) AS missing_Weight,
        sum(CASE WHEN Medal IS NULL OR Medal = '' OR Medal = 'NA' THEN 1 ELSE 0 END) AS missing_Medal
    FROM athlete_events
)
SELECT 
    total_lignes_athlete_events,
    missing_Age,
    missing_Height,
    missing_Weight,
    missing_Medal,
    (missing_Age / total_lignes_athlete_events)*100 AS missing_age_ratio,
    (missing_Height / total_lignes_athlete_events)*100 AS missing_height_ratio,
    (missing_Weight / total_lignes_athlete_events)*100 AS missing_weight_ratio,
    (missing_Medal / total_lignes_athlete_events)*100 AS missing_medal_ratio
FROM counts
```

```sql
SELECT 
       sum(CASE WHEN NOC IS NULL OR NOC = '' OR NOC = 'NA' THEN 1 ELSE 0 END) AS missing_NOC,
       sum(CASE WHEN region IS NULL OR region = '' OR region = 'NA' THEN 1 ELSE 0 END) AS missing_region,
       sum(CASE WHEN Notes IS NULL OR Notes = '' OR Notes = 'NA' THEN 1 ELSE 0 END) AS missing_Notes
FROM noc_regions
```

```sql
with counts_noc_regions AS (
    SELECT 
        count(*) AS total_lignes_noc_regions,
        sum(CASE WHEN region IS NULL OR region = '' OR region = 'NA' THEN 1 ELSE 0 END) AS missing_region,
        sum(CASE WHEN Notes IS NULL OR Notes = '' OR Notes = 'NA' THEN 1 ELSE 0 END) AS missing_Notes
    FROM noc_regions
)
SELECT 
    total_lignes_noc_regions,
    missing_region,
    missing_Notes,  
    (missing_region / total_lignes_noc_regions)*100 AS missing_region_ratio,
    (missing_Notes / total_lignes_noc_regions)*100 AS missing_Notes_ratio
FROM counts_noc_regions
```

## Duplicates

Que ce soit en considérant la colonne **ID toute seule** ou en considérant la liste **(ID,Games,Event)**, on trouve des doublons dans la table athlete_events.
La répétition des valeurs de ID est normale : un même athlète peut participer à plusieurs éditions des Jeux Olympiques et à plusieurs épreuves. Le grain de la table étant la participation à une épreuve, plusieurs lignes peuvent référencer le même athlète

```sql
SELECT ID,
count(*) AS ID_dupliques
FROM athlete_events
GROUP BY ID
HAVING count(*)>1
ORDER BY ID_dupliques DESC
```

```sql
SELECT count(*) AS total_doublons,
ID,
Games,
Event
FROM athlete_events
GROUP BY ID, Games, Event
HAVING count(*)>1
ORDER BY total_doublons DESC
```
